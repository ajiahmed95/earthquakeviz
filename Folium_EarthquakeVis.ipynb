# Earthquake data visualization from csv file

## Import package

```python
import pandas as pd
import csv
import re
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
import folium
```

## Read csv file data to DataFrame

```python
df = pd.read_csv("C:/Users/Ardi/scrapy/bmkg_di/spiders/earthquake_data.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df['Lintang - Bujur'])
```

## Cleaning Data (Lat-Long, Depth, Date) for future analysis

```python
pattern = r'^\s*([\d\.]+)\s+(LS|LU)\s+([\d\.]+)\s+BT\s*$'
df[['Latitude', 'LS_LU', 'Longitude']] = df['Lintang - Bujur'].str.extract(pattern)

# convert latitude values in LS format to negative
df.loc[df['LS_LU'] == 'LS', 'Latitude'] = '-' + df.loc[df['LS_LU'] == 'LS', 'Latitude']

# drop the intermediate column 'LS_LU'
df.drop(columns=['LS_LU'], inplace=True)

# drop the km from depth to change to numeric data
df['Depth_km'] = df['Kedalaman'].str.replace(' Km', '').astype(float)

# change date to datetype
df['Date'] = pd.to_datetime(df['Waktu Gempa'], format='%d/%m/%Y')

# change to numeric
df['Latitude'] = pd.to_numeric(df['Latitude'])
df['Longitude'] = pd.to_numeric(df['Longitude'])

print(df.head())
print(df.info())
```

## Showing lat-long data on scatter plot

```python
plt.scatter(x=df.Longitude, y=df.Latitude)
```

## Create new column for point data (for future analysis)

```python
df['geometry'] = df.apply(lambda x: Point(x.Longitude, x.Latitude), axis = 1)
print(df.head())
```

## Create GeoDataFrame using epsg:4326 coordinate georeference

```python
df_geo = gpd.GeoDataFrame(df, crs = {'init': 'epsg:4326'}, geometry = df.geometry)
print(df_geo.head())
```

## Start visualization using folium package (coordinate = center of Indonesia)

```python
m = folium.Map(location = [-2.600029, 118.015776], zoom_start = 4)

for row in df_geo.iterrows():
    row_values = row[1]
    location = [row_values['Latitude'], row_values['Longitude']]
    popup = ('Waktu Gempa: ' + str(row_values['Waktu Gempa']) + '; ' + 'Magnitudo: ' + str(row_values['Magnitudo']) + '; ' + 'Kedalaman: ' + str(row_values['Kedalaman']) + '; ' + 'Dirasakan: ' + str(row_values['Dirasakan (Skala MMI)']))
    marker = folium.Marker(location = location, popup = popup)
    marker.add_to(m)
    
display(m)
```

## Visualization the radius of earthquake based on magnitude (not following standardization)

```python
m2 = folium.Map(location = [-2.600029, 118.015776], zoom_start = 4)

for row in df_geo.iterrows():
    row_values = row[1]
    location = [row_values['Latitude'], row_values['Longitude']]
    popup = ('Waktu Gempa: ' + str(row_values['Waktu Gempa']) + '; ' + 'Magnitudo: ' + str(row_values['Magnitudo']) + '; ' + 'Kedalaman: ' + str(row_values['Kedalaman']) + '; ' + 'Dirasakan: ' + str(row_values['Dirasakan (Skala MMI)']))
    marker = folium.CircleMarker(location = location, 
                                 popup = popup, 
                                 radius = int(row_values['Magnitudo'])*5,
                                 weight = 1,
                                 color = 'red',
                                 opacity = 0.3,
                                 fill_color = 'red',
                                 fill_opacity = 0.1)
    marker.add_to(m2)

   
    
display(m2);
```
